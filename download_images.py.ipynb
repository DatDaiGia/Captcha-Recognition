```python
import request
import time
import os
```

```python
url = "https://www.e-zpassny.com/vector/jcaptcha.do"
total = 0
num_images = 500
```

```python
for i in range(0, num_images):
    try:
        r = request.GET(url)
        p = os.path.join("ouput/", "{}.png".format(str(total).zfill(5)))
        f = open(p, "w")
        f.write(r.content)
        f.close()
        
        print("[INFO] downloaded: {}".format(p))
        total += 1
    except:
        print("[INFO] error downloading image...")
    
    time.sleep(0.1)
```
